```python
import cv2
import time
import os
from datetime import datetime

data_file_name = "dataset-0-0-0"
if not os.path.exists(data_file_name):
    os.makedirs(data_file_name)

video_path = r"C:\Users\zbook 17 g3\Desktop\CV PROJECTS\New folder\29.08.2023_18.49.20_REC.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Could not open video file: {video_path}")
else:
    while True:
        ret, img = cap.read()
        if not ret:
            print("Video stream ended.")
            break

        cv2.imshow("frames", img)

        key = cv2.waitKey(1)
        if key == ord("s"):
            timee = datetime.now().strftime("%Y%m%d%H%M%S")
            screen_name = os.path.join(data_file_name, f"screen_{timee}.png")
            cv2.imwrite(screen_name, img)
            print("Image has been captured")

        if key == ord('q'):
            print("Process is done")
            break

    cap.release()
    cv2.destroyAllWindows()
```
